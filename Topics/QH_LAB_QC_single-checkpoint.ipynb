# Explicación del circuito cuántico

### QuantumCircuit(1)

Esta línea crea un **circuito cuántico** con 1 qubit.

- Si quisieras medir o interactuar con el qubit, necesitarías también un **registro clásico**.  
- En este caso, el registro clásico no está explícitamente creado, así que por defecto **Qiskit** lo maneja internamente si agregas mediciones.

### qc.draw("mpl")

Este método permite **visualizar el circuito** en un diagrama de estilo gráfico usando **Matplotlib**.

# Proceso de Acoplamiento y `transpile`

### coupling_map

- Define el **mapa de acoplamiento** entre qubits en un `backend` físico.  
- Cada sublista `[i, j]` indica que el qubit `i` puede interactuar con el qubit `j`.  
- Ejemplo: `[[0]]` indica que solo hay un qubit y no hay conexiones adicionales.  
- Función adicional:
  - Permite al **transpiler** ajustar el circuito a la topología del hardware.

---

### measure_all

- Agrega operaciones de **medición a todos los qubits** del circuito.  
- Los resultados se almacenan en un **registro clásico** creado automáticamente.  
- Opciones adicionales:
  - `register=None` → permite especificar un registro clásico particular.  
  - `inplace=True/False` → si se modifica el circuito original o se crea una copia.  
- Función extra: prepara el circuito para **obtener resultados de simulación o ejecución en hardware**.

---

### generate_preset_pass_manager

- Genera un **gestor de passes predefinido** para optimizar y transpilar circuitos.  
- Ajusta el circuito para cumplir restricciones de **hardware** y optimizar eficiencia.  
- Opciones importantes:
  - `optimization_level=0..3` → controla cuán agresivas son las optimizaciones.  
  - Permite incluir/excluir ciertos tipos de pases, como fusión de puertas o reordenamiento de qubits.  
- Función extra: combina varios **pases de optimización y mapeo de qubits** en un solo gestor.

---

### run (del pass manager)

- Aplica el **pass manager** al circuito `qc` y devuelve el **circuito transpileado**.  
- Ajusta:
  - Las mediciones agregadas.  
  - El circuito al **coupling_map** proporcionado.  
  - La profundidad y eficiencia del circuito.  
- Función adicional: asegura que el circuito sea **compatible con backends específicos** y optimizado para ejecución.

---

### draw("mpl")

- Muestra el **diagrama final del circuito transpileado** usando Matplotlib.  
- Permite visualizar:
  - Mediciones agregadas.  
  - Ajustes realizados por el pass manager.  
  - Optimización de puertas y reducción de profundidad.  
- Opciones adicionales:
  - `"text"` → diagrama ASCII.  
  - `"latex"` → diagrama en LaTeX.  
  - `"latex_source"` → código LaTeX para personalización avanzada.
- Función extra: útil para **presentaciones educativas** y revisión visual de la estructura del circuito.

---

### Notas adicionales

- El **registro clásico** ahora contiene los resultados de la medición de todos los qubits.  
- La **transpilación** permite adaptar el circuito a restricciones físicas del hardware cuántico.  
- El **pass manager** asegura eficiencia y compatibilidad tanto en simuladores como en backends reales.



````python
# Escribir el mapa de acoplamiento [[0,1],...]
coupling_map = [[0]]
qc.measure_all()
# Transpilar el `qc` using el `generate_preset_pass_manager` y la lista de `coupling_map`
pm = generate_preset_pass_manager()
qc_transpiled = pm.run(qc)

qc_transpiled.draw("mpl")
````

# Primitiva `Sampler` y Activacvión de AerSimulator

### AerSimulator

- Crea un **backend de simulación** para ejecutar circuitos cuánticos localmente.  
- Simula la **ejecución de qubits** y las mediciones como si fuera un backend real.  
- Funciones adicionales:
  - Permite simular **ruido, decoherencia y errores de medición**.  
  - Soporta múltiples **shots** para obtener distribuciones estadísticas.

---

### Sampler

- Configura un **sampler** para ejecutar circuitos en el backend de simulación.  
- Funciones adicionales:
  - Ejecuta varios circuitos de manera **eficiente en batch**.  
  - Genera **distribuciones de probabilidad** de los estados medidos.

---

### generate_preset_pass_manager

- Crea un **gestor de pases predefinido** adaptado al backend y nivel de optimización.  
- Funciones adicionales:
  - Optimiza el circuito según la topología y restricciones del backend.  
  - Reduce la **profundidad del circuito** y la cantidad de puertas.  
  - Permite ajustar el nivel de optimización (`0`=mínima, `3`=máxima).

---

### run del sampler

- Envía el circuito transpileado al **sampler** para su ejecución.  
- Funciones adicionales:
  - Permite ejecutar **varios circuitos a la vez**.  
  - Devuelve un **job object**, que puede ser monitoreado o esperado hasta completarse.

---

### job.result

- Obtiene los **resultados de la ejecución** del sampler y encapsula toda la información de los circuitos ejecutados.  
- El resultado es un **objeto de resultados**, que contiene para cada circuito un **diccionario detallado**.  
- Estructura típica del diccionario de salida (`job.result()[i].data.meas`):
  - `counts` → diccionario con los conteos de cada estado medido (por ejemplo, `{'0': 512, '1': 488}`).  
  - `probabilities` → distribución de probabilidad de cada estado (por ejemplo, `{'0': 0.512, '1': 0.488}`).  
  - `shots` → número total de ejecuciones realizadas.  
  - `memory` → lista de resultados individuales de cada ejecución (opcional, según backend).  
  - `metadata` → información adicional sobre la ejecución, como tiempo de simulación y configuración del backend.  
- Funciones adicionales:
  - Permite acceder a los **conteos** con `.get_counts()`.  
  - Permite analizar **estadísticas** del circuito y probabilidades para cada estado.  
  - Útil para comparar **resultados teóricos vs simulación**.

---

### get_counts

- Extrae directamente los **conteos de resultados** de la medición para cada estado del qubit.  
- Funciones adicionales:
  - Devuelve un **diccionario** con claves como `'0'`, `'1'` y valores indicando cuántas veces se midió cada estado.  
  - Permite analizar la **distribución probabilística** de la superposición de los qubits.

---

### Flujo de trabajo

- Este flujo combina **simulación**, **optimización** y **medición** de circuitos cuánticos.  
- `AerSimulator` junto con `Sampler` permite **experimentar con circuitos complejos sin hardware físico**.  
- La combinación con un **pass manager** asegura que los circuitos sean eficientes y adaptables a backends reales o simulados.  
- El **job.result()** es la fuente principal de información para **interpretar los resultados** de la ejecución del circuito.


````python
backend = AerSimulator()

# Set up the sampler
sampler = Sampler(mode=backend)

# Submit the circuit to Sampler
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
job = sampler.run(pm.run([qc]))

# Get the results
results_sampler = job.result()

print(results_sampler[0].data.meas.get_counts())
````

## Primitiva Estimator

### ¿Qué es `results_estimator`?

- Es el **objeto de resultados** devuelto por el `Estimator` después de ejecutar un circuito y evaluarlo sobre una lista de observables.  
- Se obtiene con:  
  > results_estimator = job.result()

---

### Contenido típico de `results_estimator`

El objeto contiene información estructurada en forma de **diccionario interno**.  
Para un circuito con observables `[Z, X, Y, I]`, la salida podría verse así:

> results_estimator.values  
Ejemplo de salida:  
> [0.98, 0.01, 0.00, 1.00]

- Cada valor corresponde al **valor esperado (expectation value)** de un observable:
  - `⟨Z⟩ ≈ 0.98`
  - `⟨X⟩ ≈ 0.01`
  - `⟨Y⟩ ≈ 0.00`
  - `⟨I⟩ = 1.00`

Otros atributos importantes:  
- > results_estimator.metadata  
  Información extra sobre la ejecución (backend, shots, etc.).  
- > results_estimator.quasi_dists  
  Distribuciones aproximadas de probabilidad asociadas (en versiones recientes de Qiskit).  

---

### Interpretación

- Los valores devueltos están en el rango **[-1, 1]** porque provienen de operadores de Pauli.  
- Ejemplo:
  - Si el qubit está en el estado inicial |0⟩, el valor esperado de `Z` será cercano a **+1**, mientras que `X` y `Y` serán cercanos a **0**.  
  - El observable identidad `I` siempre devuelve **1** porque no altera el estado.

---

### Cómo graficar `results_estimator`

Podemos representar los valores esperados en un **gráfico de barras**:

> import matplotlib.pyplot as plt  
>   
>-------- Valores esperados------------------  
> values = results_estimator.values  
> labels = ["Z", "X", "Y", "I"]  
>   
> plt.bar(labels, values)  
> plt.ylim(-1, 1)  
> plt.title("Expectation values from Estimator")  
> plt.xlabel("Observables")  
> plt.ylabel("⟨O⟩")  
> plt.axhline(0, color="black", linewidth=0.8)  
> plt.show()  

---

### Visualización esperada

- El gráfico mostrará una barra para cada observable (`Z, X, Y, I`).  
- Altura de la barra = valor esperado ⟨O⟩ del observable.  
- Esto permite **interpretar el estado cuántico** en función de cómo responde a distintos operadores de Pauli.

---

### Notas adicionales

- Este tipo de análisis es esencial en algoritmos **variacionales** (VQE, QAOA).  
- Graficar los valores esperados ayuda a **comparar el estado cuántico real vs esperado**.  
- Para sistemas con más qubits, los observables pueden ser combinaciones como `"ZZ"`, `"XX"`, etc., y el gráfico se extiende a esas combinaciones.


````python
# Set up the Estimator
estimator = Estimator(mode=backend)

# Define some observables
Z = SparsePauliOp("I")
X = SparsePauliOp("X")
Y = SparsePauliOp("Y")
I = SparsePauliOp("I")
observables = [Z, X, Y, I]


# Submit the circuit to Estimator
pub = (qc, observables)
job = estimator.run(pubs=[pub])

# Get the results
results_estimator = job.result()

# Show Values
exp_values = results_estimator[0].data.evs
print(f"Expectation values: {list(zip(observables_list, exp_values))}")
````

## Optimización y Transpilación del circuito

###  pm = generate_preset_pass_manager(backend=backend, optimization_level=1)

**Función principal:**  
- `generate_preset_pass_manager()` crea un **administrador de pasos de compilación (pass manager)** para transpilar el circuito cuántico.  
- La **transpilación** adapta tu circuito lógico a las **restricciones físicas del backend** (topología, puertas disponibles, tiempos de ejecución).  

**Parámetros clave:**  
- `backend=backend` → especifica el dispositivo cuántico o simulador donde se ejecutará el circuito.  
- `optimization_level=1` → nivel de optimización (va de 0 a 3):  
  - **0** → sin optimización, solo mapeo básico.  
  - **1** → optimización ligera, balance entre velocidad y calidad.  
  - **2** → optimización media, mayor reducción de compuertas y profundidad.  
  - **3** → optimización agresiva, máxima reducción del circuito (puede tardar más).  

---

###  qc_transpiled = pm.run(qc)

**Función principal:**  
- Aplica el **pass manager** al circuito original `qc`.  
- Devuelve un **circuito transpileado (`qc_transpiled`)**, compatible con el backend elegido y con optimizaciones según el nivel seleccionado.  

**Funciones adicionales del objeto transpileado:**  
- `qc_transpiled.depth()` → devuelve la profundidad del circuito (número de capas de puertas).  
- `qc_transpiled.size()` → número total de operaciones cuánticas.  
- `qc_transpiled.count_ops()` → cuenta cuántas veces aparece cada puerta cuántica.  

---

###  qc_transpiled.draw("mpl")

**Función principal:**  
- Dibuja el circuito ya **optimizado y transpileado** en formato gráfico con **Matplotlib**.  

**Otras opciones de visualización:**  
- `"text"` → representación en ASCII.  
- `"latex"` → diagrama en LaTeX (ideal para notebooks).  
- `"latex_source"` → devuelve el código LaTeX que genera el circuito.
