# Superposición y Estados Puros en Computación Cuántica

En computación cuántica, el estado de un sistema cuántico (por ejemplo, un qubit) se describe mediante un **vector en un espacio de Hilbert**. Para un solo qubit, el estado general se escribe como:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

donde:

- $ |0\rangle $ y $ |1\rangle $ son los **estados base** (análogos a 0 y 1 clásicos),
- $ \alpha $ y $ \beta $ son **números complejos** (amplitudes),
- y deben cumplir la **condición de normalización**:

$$
|\alpha|^2 + |\beta|^2 = 1
$$

Esto asegura que la probabilidad total de medir algún estado sea 1.

---

## Superposición

La **superposición** es una propiedad fundamental de la mecánica cuántica: un qubit puede estar en una combinación de los estados base al mismo tiempo.

Ejemplo clásico de superposición:

$$
|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)
$$

Esto significa que el qubit tiene **50% de probabilidad** de colapsar a $ |0\rangle $ y **50%** a $ |1\rangle $ si lo medimos.

---

## Estado puro vs estado mixto

- Un **estado puro** es un estado cuántico que se puede describir completamente mediante un vector de estado $ |\psi\rangle $.
- Un **estado mixto** es una colección estadística de estados puros, y requiere el uso de una **matriz densidad** para su representación.

En este laboratorio, solo trabajaremos con **estados puros**, los cuales podemos manipular y visualizar directamente con vectores de estado.

---

## Visualización de superposiciones

La **esfera de Bloch** nos permite visualizar los estados puros de un qubit como puntos sobre la superficie de una esfera.

Por ejemplo:

- $ |0\rangle $: Polo norte
- $ |1\rangle $: Polo sur
- $ \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) $: Eje X positivo
- $ \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle) $: Eje X negativo
- $ \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle) $: Eje Y positivo
- $ \frac{1}{\sqrt{2}}(|0\rangle - i|1\rangle) $: Eje Y negativo

Estas posiciones corresponden a diferentes **superposiciones con fase relativa**, que afectan el resultado de mediciones y operaciones cuánticas.

---

## Resumen

- Los qubits pueden estar en superposición de $ |0\rangle $ y $ |1\rangle $.
- Esta superposición se representa con amplitudes $ \alpha $ y $ \beta $ que cumplen $ |\alpha|^2 + |\beta|^2 = 1 $.
- Estados puros se representan como vectores $ |\psi\rangle $; los veremos visualmente en la esfera de Bloch.

En computación cuántica, la **puerta X** es una de las puertas más fundamentales. Es el **análogo cuántico de la compuerta NOT clásica**.

---

# ¿Qué hace?

Transforma el estado base $\ket{0}$ en $\ket{1}$, y viceversa:

$$
X\ket{0} = \ket{1}, \quad X\ket{1} = \ket{0}
$$

---

# Representación matemática

La puerta X se representa por la **matriz de Pauli-X**:

$$
X =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$


````python
# Estado |1⟩
qc_1 = QuantumCircuit(q_reg, c_reg)
qc_1.x(0)
state_1 = Statevector.from_instruction(qc_1)
print(state_1)
````

En computación cuántica, la **puerta Hadamard** es fundamental porque permite crear **superposición cuántica**. Es una de las puertas más importantes en algoritmos como el de Grover, Deutsch-Jozsa, y muchos más.

---

# ¿Qué hace?

La puerta Hadamard transforma los estados base $\ket{0}$ y $\ket{1}$ en estados superpuestos:

$$
H\ket{0} = \frac{1}{\sqrt{2}}(\ket{0} + \ket{1}) = \ket{+} \\
H\ket{1} = \frac{1}{\sqrt{2}}(\ket{0} - \ket{1}) = \ket{-}
$$

---

# Representación matemática

La puerta Hadamard se representa mediante la siguiente matriz:

$$
H =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}
$$

````python
# Estados en Superposición
# Estado (|0⟩ + |1⟩)/√2 usando puerta Hadamard

qc_plus = QuantumCircuit(q_reg, c_reg)
qc_plus.h(0)
state_plus = Statevector.from_instruction(qc_plus)
print(state_plus)
````

# La Esfera de Bloch

La **esfera de Bloch** es una herramienta visual fundamental para representar el estado de un **qubit** en **computación cuántica**.

---

## ¿Qué es?

Un qubit puede estar en un estado base como:

- $$|0\rangle = \begin{bmatrix}1 \\ 0\end{bmatrix}$$
- $$|1\rangle = \begin{bmatrix}0 \\ 1\end{bmatrix}$$

...pero también puede estar en una **superposición** de estos dos estados:

$$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle
$$

con $ \alpha, \beta \in \mathbb{C} $ y $ |\alpha|^2 + |\beta|^2 = 1 $
(normalización).

Gracias a esta restricción, y descartando la **fase global**, todos los estados puros de un qubit pueden representarse como puntos sobre la **superficie de una esfera unitaria**: la **esfera de Bloch**.

---

## Forma general del estado

Cualquier estado puro de un qubit se puede expresar como:

$$
|\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + e^{i\phi} \sin\left(\frac{\theta}{2}\right)|1\rangle
$$

Donde:

- $ \theta \in [0, \pi] $: ángulo polar
- $ \phi \in [0, 2\pi] $: ángulo azimutal

---

## Ejes de la esfera

- **Eje Z**:
  - Polo norte: $ |0\rangle $
  - Polo sur: $ |1\rangle $

- **Eje X**:
  - $ \frac{|0\rangle + |1\rangle}{\sqrt{2}} $

- **Eje Y**:
  - $ (|0\rangle + i|1\rangle)/\sqrt{2} $

---

## Transformación de coordenadas esféricas (Esfera de Bloch)

En la representación de la esfera de Bloch, cualquier estado puro de un **qubit** puede expresarse mediante dos ángulos esféricos:

- $\theta \in [0, \pi]$: ángulo polar (vertical)
- $\phi \in [0, 2\pi)$: ángulo azimutal (horizontal)

El estado cuántico en notación de Dirac se escribe como:

$$
|\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + e^{i\phi} \sin\left(\frac{\theta}{2}\right)|1\rangle
$$

---

### Conversión a coordenadas cartesianas $(x, y, z)$

Para representar este estado en la esfera de Bloch, usamos la siguiente transformación:

$$
\begin{aligned}
x &= \sin(\theta) \cos(\phi) \\
y &= \sin(\theta) \sin(\phi) \\
z &= \cos(\theta)
\end{aligned}
$$

Estas coordenadas determinan la **posición del vector** que representa el qubit en la esfera de Bloch.

---

````python
# |0⟩

theta = 1
phi = 2

x = np.sin(theta) * np.cos(phi)
y = np.sin(theta)
z = np.cos(theta)
````

## Normalización de estados cuánticos

En mecánica cuántica, el estado de un sistema se describe por un vector de estado $|\psi\rangle$ en un espacio de Hilbert. Para que este vector represente un estado físico válido, **debe estar normalizado**.

---

### ¿Qué significa normalizar un estado?

Normalizar un estado significa que su norma (o longitud) debe ser 1:

$$
\langle \psi | \psi \rangle = 1
$$

Esto asegura que las probabilidades asociadas a las mediciones del sistema suman 1, lo cual es un requisito fundamental en la interpretación probabilística de la mecánica cuántica.

---

### Estado genérico de un qubit

Un estado arbitrario de un qubit se escribe como:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

donde $\alpha, \beta \in \mathbb{C}$ son amplitudes complejas.

La condición de normalización implica que:

$$
|\alpha|^2 + |\beta|^2 = 1
$$

---

### ¿Por qué es importante?

- Las cantidades $|\alpha|^2$ y $|\beta|^2$ representan las probabilidades de encontrar el sistema en los estados base $|0\rangle$ y $|1\rangle$ respectivamente.
- Si el estado no está normalizado, estas "probabilidades" no sumarían 1, violando las reglas básicas de probabilidad.

---

### Normalizar un vector

Si tienes un vector no normalizado

$$
|\psi'\rangle = a|0\rangle + b|1\rangle
$$

puedes normalizarlo dividiendo por su norma:

$$
|\psi\rangle = \frac{|\psi'\rangle}{\sqrt{|a|^2 + |b|^2}} = \frac{a|0\rangle + b|1\rangle}{\sqrt{|a|^2 + |b|^2}}
$$

---

### En Qiskit

Cuando defines un estado usando amplitudes, Qiskit espera que ya estén normalizadas para representar un estado físico válido.

````python
# Comprobación de Normalización

norm = np.sum(np.abs(state_0.data) ** 2)
print(norm)

norm_1 = np.sum(np.abs(state_1.data) ** 2)
print(norm)
````
